# Three-Node Chain Example

As per review: minimal three-node chain with explicit computations.

```
import networkx as nx
import numpy as np

# Three-node chain
G3 = nx.path_graph(3)
L3 = nx.laplacian_matrix(G3).toarray().astype(float)
eigvals3, eigvecs3 = np.linalg.eigh(L3)
eigvals3[eigvals3 < 1e-10] = 1.0
L3_plus = eigvecs3 @ np.diag(1.0 / eigvals3) @ eigvecs3.T

# Coherence weights
G_edge3 = {(0,1): 0.3, (1,0): -0.3, (1,2): 0.1, (2,1): -0.1}

# Path [0,1,2]
path = [0,1,2]
C = G_edge3[(0,1)] + G_edge3[(1,2)]
AD = L3_plus[0,0] + L3_plus[2,2] - 2*L3_plus[0,2]
AVID_ex = AD * np.exp(-C) * 1.0  # Toy A_inf=1
print('AVID for three-node:', AVID_ex)
```
