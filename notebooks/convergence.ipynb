# ∞-Algebra Convergence Simulator

This notebook implements the convergence check for the Infinity Algebra sector as per VID paper Proposition 2.3.

```
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Dodecahedral graph
G = nx.dodecahedral_graph()
L = nx.laplacian_matrix(G).toarray().astype(float)
eigvals, eigvecs = np.linalg.eigh(L)
eigvals[0] = 1.0
L_plus = eigvecs @ np.diag(1.0/eigvals) @ eigvecs.T

H_dim = L_plus.shape[0]
d_inf = np.eye(H_dim) - 0.05 * L_plus

def O_k(k):
    decay = 1.0 / (k+1)**2.2
    return decay * np.random.randn(H_dim, H_dim)

N_terms = 100
norms = []
errors = []
O_cum = np.zeros((H_dim, H_dim))

for k in range(1, N_terms+1):
    O_cum += O_k(k)
    norms.append(np.linalg.norm(O_k(k), ord=2))
    H_N = d_inf @ O_cum @ d_inf
    if k > 1:
        err = np.linalg.norm(H_N - H_prev)
        errors.append(err)
    H_prev = H_N.copy()

# Plots
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.semilogy(norms, 'o-')
plt.title('‖O_k‖ → 0')
plt.xlabel('grade k')
plt.grid()

plt.subplot(1,2,2)
plt.semilogy(errors, 'r.-')
plt.title('‖H^(N+1)∞ - H^(N)∞‖ → 0')
plt.xlabel('truncation N')
plt.axhline(1e-12, color='k', ls='--', label='machine precision')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

print(f'Final norm of H_∞ ≈ {np.linalg.norm(H_N):.6f}')
```
